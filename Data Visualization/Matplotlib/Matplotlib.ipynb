## Matplotlib

Most popular plotting library for Python.<br>
Similar feel to MATLAB graphical plotting.



```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline 
#see plot in jupyter
#plt.show() outside jupyter
```

```python
x = np.linspace(0,5,11)
y = x**2
```

#### Functional Method

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.show()
```

#### Object-Oriented Plotting

```python
fig = plt.figure()  # BLANK CANVAS
axes = fig.add_axes([0.1,0.1,0.8,0.8])
# LIST CONTAINS 4 elements
# LEFT of axes , BOTTOM of axes , WIDTH , HEIGHT
# RANGE FROM 0-1 : PERCENT OF CANVAS
```

```python
fig = plt.figure()  
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Fig1')
axes.plot(x,y)
```

```python
fig1 = plt.figure()
axes1 = fig1.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig1.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
plt.show()
```

Create Subplots

We manually created canvas/axes above.<br>
Using <code>.subplot()</code> we can create 'array' of axes.

```python
fig,axes = plt.subplots(nrows=1,ncols=2) #sort of Tuple Unpacking
axes[0].plot(x,y) # Equal to add_axes() above
axes[1].plot(y,x)
axes[0].set_title('First Plot')
axes[1].set_title('SecondPlot')

plt.tight_layout() # prevents overlapping
```

### Figure Size and DPI

```python
fig = plt.figure(figsize=(6,6))
axes = fig.add_axes([0,0,1,1])
# figure size is in inches
# dpi - dots per inches
```

```python
fig,axes = plt.subplots(figsize=(8,2),nrows = 1 , ncols = 2)
axes[0].plot(x,y)
axes[0].set_title('Fig1')
axes[1].plot(y,x)
axes[1].set_title('Fig2')
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,label='Squared')
axes.plot(x,x**3,label='Cubed')

axes.legend()
plt.show()
```

**Plot Appearance**

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'white',linewidth = 1,alpha = 0.3,\
       linestyle='--') 
# RGB codes also applicable
# alpha controls transparency of line
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='white',marker='o',markersize=10,\
        markerfacecolor='red')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='white',marker='o',markersize=10,\
        markerfacecolor='red')
ax.set_xlim([0,2])
ax.set_ylim([0,2])
plt.show()
```
