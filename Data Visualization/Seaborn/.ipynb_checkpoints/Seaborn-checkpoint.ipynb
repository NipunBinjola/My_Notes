### Distribution Plots

Allows us to show distribution of univariate set of observations.<br> 

```python
#import matplotlib as mpl
#mpl.rcParams['patch.force_edgecolor'] = True
import seaborn as sns
%matplotlib inline
#from jupyterthemes import jtplot
#jtplot.style()
```

```python
# BUILT IN DATASETS
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
# BIN SIZE
```

```python
sns.jointplot(x = 'total_bill',y='tip',data=tips )
# THINGS TO COMPARE
```

```python
sns.jointplot(x = 'total_bill',y='tip',data=tips,kind='hex' )
```

```python
sns.jointplot(x = 'total_bill',y='tip',data=tips,kind='kde' )
```

```python
sns.pairplot(data=tips,hue='sex')
#DOES JOINTPLOT FOR ALL COMBINATIONS FOR NUMERRICAL COLUMN
# HUE contains categorical feature to differentiate
```

```python
sns.rugplot(tips['total_bill'])
#DRAW A DASH FOR EVERY SINGLE POINT ON DISTRIBUTION LINE
#SIMILAR TO HSITOGRAM - IT CONTAINED 'BINS'
#??
```

```python
sns.kdeplot(tips['total_bill'])
```
